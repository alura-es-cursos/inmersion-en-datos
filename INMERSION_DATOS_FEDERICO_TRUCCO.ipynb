```python
import pandas as pd

#Leer la ruta 'path' de la BbDd del sheet.csv
inmuebles = pd.read_csv('/content/drive/MyDrive/ColabNotebooks/inmuebles_bogota.csv')

#.head() muestra 1ras 5 líneas de la tabla
inmuebles.head() 
```

```python
#Diccionario = {'key' : 'value'} => Tratamiento a caracteres especiales
columnas = {'Baños' : 'Banos', 'Área':'Area'}

#Asigno los nuevos valores a columnas
inmuebles = inmuebles.rename(columns=columnas)

#.sample(n°) = n° de muestras q deseo tomar de BbDd
inmuebles.sample(10)
```

```python
#.str.split(expand=True) = Convierte el pandas series en dataframes

inmuebles.Valor.str.split(expand=True)
```

```python
#Creo nueva variable 'valor'
valor = inmuebles.Valor.str.split(expand=True)

#Genero nueva columna 'Moneda' cuyo valor = columna [0] de split anterior
inmuebles['Moneda'] = valor[0]

#Genero nueva columna 'Precio' cuyo valor = columna [1] de split anterior
inmuebles['Precio'] = valor[1]

#.sample(3) = Tomo muestra de 3 elementos
inmuebles.sample(3)
```

```python
#Verifico de q tipo son las columnas nuevas creadas

inmuebles.info()
```

```python
#Si se intentan sumar los object los concatena xq son string

inmuebles['Precio'].sum()
```

```python
#.str.replace('.','') = Para quitar/reemplazar el punto (.) del precio

inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','')
```

```python
#.str.replace('.','', regex=True) = Para evitar el warning de arriba

inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','', regex=True)
```

```python
#[['Precio','Barrio']] = Para q pandas tome los datos como dataframe

inmuebles[['Precio','Barrio']]
```

```python
#Verifico si al sacar el ($) y el (.) ya los toma como n°

inmuebles.info()
```

```python
#Creo nueva columna 'Precio_Millon'
#Precio.astype('float') = Transforma de object (string) a float 
#/1000000 = le quito 6 ceros al precio

inmuebles['Precio_Millon'] = inmuebles.Precio.astype('float')/1000000

#Verifico los cambios
inmuebles.info()
```

```python
#.sample(10): Muestra aleatoria de 10 filas del DataFrame 
inmuebles.sample(10)

#.Tipo: muestra la columna 'Tipo' del DataFrame
inmuebles.Tipo

#Creo DataFrame 'tres_inmuebles' q contiene las filas del DataFrame original 'inmuebles' y contiene alguno de los tres valores: 'Apartamento', 'Casa' o 'Bodega
tres_inmuebles = inmuebles[inmuebles.Tipo.isin(['Apartamento','Casa','Bodega'])]

tres_inmuebles
```

```python
plt.figure(figsize=(10,6))

#Establece estilo blanco con cuadrícula
sns.set_style('whitegrid') 

#Define una paleta de colores
palette = sns.color_palette(['#8e44ad', '#f1c40f', '#c0392b'])

#Uso la paleta de colores primarios
grafica_tres_inmuebles = sns.histplot(data=tres_inmuebles, x='Precio_Millon', kde=True, hue='Tipo', palette=palette)

grafica_tres_inmuebles.set_title('Precio de Apartamentos, Casas, Bodegas', fontsize=15)

plt.xlim((50,1000))
plt.ylim(0,750)

grafica_tres_inmuebles.set_xlabel('Precio_Millon', fontsize=12)
grafica_tres_inmuebles.set_ylabel('Count', fontsize=12)

plt.show()


#'\033[1m' + palabra = Aplica letra negrita
#palabra + '\033[0m' = Quita letra negrita

#'\033[3m' + palabra = Aplica letra cursiva
#palabra + '\033[0m' = Quita letra cursiva

#Agrego la conclusión
print('\033[1m' + "CONCLUSIÓN: " + '\033[0m' '\033[3m' + "En el gráfico anterior se puede apreciar cómo la distribución de valor de los inmuebles de Bogotá se divide entre:")
print("apartamentos y casas con un valor superior a 200 millones de pesos colombianos.")
print("Por otro lado, el valor de las bodegas es significativamente inferior a estos inmuebles, con un valor máximo de alrededor de 150 millones de pesos colombianos.")
```

```python
inmuebles_barrio_m2 = inmuebles.groupby('Barrio').mean().head(15)

#Calculo 'Precio m2'
inmuebles_barrio_m2['Precio_m2'] = inmuebles_barrio_m2['Precio_Millon'] / inmuebles_barrio_m2['Area']

#ascending=True = Gráfica se ordena de manera ascendente
#inplace=True = Ordena las barras de menor a mayor
inmuebles_barrio_m2.sort_values(by='Precio_m2', ascending=True, inplace=True)

plt.figure(figsize=(10 ,6))

#.barplot = Gráfico de barras
#palette='RdYlGn' = Va del ROJO al VERDE
#orient='h' = Se orienta HORIZANTALMENTE
grafica = sns.barplot(data=inmuebles_barrio_m2, y=inmuebles_barrio_m2.index, x='Precio_m2', palette='RdYlGn',orient='h')

#color='darkgreen' = Cambia el color de la letra
#fontsize=15 = Cambia el tamaño de la letra
grafica.set_title('Valor m2 por barrio', color='darkgreen', fontsize=15)
grafica.set_xlabel('$ millones', color='darkgreen', fontsize=12)
grafica.set_ylabel('Barrio', color='darkgreen', fontsize=12)

plt.show()
```
